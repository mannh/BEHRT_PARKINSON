```python
import pandas as pd
```

```python
df = pd.read_csv('summay.csv').rename(columns={'speech':'Speech'})
```

```python
df.drop(columns=['MAE']).plot(x='Predicted Visit', figsize=(20,10),title='ALS Functions MAE', ylabel='MAE', grid=True)
```

```python
df.plot(x='Predicted Visit', y='MAE', figsize=(20,10),title='ALSFRS MAE', ylabel='MAE', grid=True)
```
